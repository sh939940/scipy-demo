```python
import scipy as sc
import numpy as np


# Define the Function
def f(x):
  return x**2+10*np.sin(x)

# Use fminbound to Find the Minimum
a=sc.optimize.fminbound(
    f,
    x1=-10,
    x2=10,
    xtol=0.01,
    full_output=True, # return more than just the x value
    disp=3   # display detailed output during optimization
)

print(a)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sc

# Function Definition
def f(x):
  return x**2+10*np.sin(x)

# Visualize the Function
x = np.arange(-10,10,0.1)  # Generates values from -10 to 10 in steps of 0.1
plt.plot (x,f(x))
plt.show()

a = sc.optimize.fmin_bfgs(
    f,
    x0=50,
    epsilon = 10, # step size used in gradient approximation
    disp =1,      # display optimization output (1 = True)
    retall= 1,     # return all intermediate results (True)
    maxiter =2   # maximum number of iterations = 2
)

print(a)
```

```python
#Use basinhopping to find the global minimum of a non-convex function with many local minima by combining random jumps with local optimization.

a=sc.optimize.basinhopping(
    f,
    x0=70,
    niter= 20,
    stepsize =.2,
    disp = 3
)

print(a)
```

```python
#This code uses SciPy's basinhopping algorithm to find the global minimum of a specific function with two variable
import scipy as sc
import numpy as np

def func2d(x):
  f=np.cos(14.5*x[0]-0.3)+(x[1]+0.2)*x[1]+(x[0]+0.2)*x[0]
  df = np.zeros(2)
  return f,df

x0 = [1.0,1.0]
ret = sc.optimize.basinhopping(
    func2d,
    x0,
    minimizer_kwargs={"method":"L-BFGS-B", "jac":True},
    #specifies the local minimization method (L-BFGS-B is a quasi-Newton method good for smooth functions)
    niter=2000)
print("global minimum: x = [%.4f, %.4f], f(x0) = %.4f" %
      (ret.x[0],ret.x[1],ret.fun))
```

```python
# This code fits a non-linear exponential function to some noisy synthetic data using SciPy's curve_fit.
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def func(x,a,b,c):
  return a*np.exp(-b*x)+c
  # a:scales the function vertically
  # b: controls the rate of exponential decay
  # c: the vertical offset (shifts the graph up or down)

xdata = np.linspace(0,4,50)
 # xdata: 50 evenly spaced points between 0 and 4.

y = func(xdata,2.5,1.3,0.5)
#Compute the true y-values based on your function with known parameters

#Add random noise to simulate real-world noisy data.
np.random.seed(1729)
y_noise = 0.2* np.random.normal(size=xdata.size) # generates noise from a normal distribution.
ydata = y + y_noise

# plot the noise data
plt.plot(xdata , ydata ,'b-', label='data')

popt, pcov = curve_fit(func, xdata, ydata) #fitting
plt.plot(xdata, func(xdata, *popt), 'r-', #draw the red line
         label='fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt))

#fit with bounds
popt, pcov =curve_fit(func,xdata,ydata,bounds=(0,[3., 1., 0.5]))
plt.plot(xdata, func(xdata, *popt),  #draw the green line
         'g--',label='fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt))

#Final Touches and Show Plot
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

```
